```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load and clean data
df = pd.read_csv("housing.csv")
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(axis=1, how='all', inplace=True)  # drop columns with all NaNs

# Select numerical columns only
numerical_df = df.select_dtypes(include=["float64", "int64"])

# Drop any remaining NaNs just for correlation analysis
correlation_matrix = numerical_df.dropna().corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Heatmap of Feature Correlations")
plt.show()

# Plot proper pairplot
sns.pairplot(numerical_df.dropna(), diag_kind="hist", corner=True)
plt.suptitle("Pair Plot of Numerical Features", fontsize=20)
plt.show()
```
